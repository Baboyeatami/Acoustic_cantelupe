```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split, cross_val_score, StratifiedKFold
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import (
    classification_report, confusion_matrix, accuracy_score,
    roc_auc_score, roc_curve
)

# ML models
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from xgboost import XGBClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier

# Load reduced dataset
df = pd.read_csv('reduced_kbest_dataset.csv')  # or 'reduced_kbest_dataset.csv' or 'reduced_rf_dataset.csv'

# Target and Features
target_column = 'Label'
X = df.drop(columns=[target_column])
y = df[target_column]

# Label encoding
le = LabelEncoder()
y = le.fit_transform(y)
class_names = le.classes_

# Scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, stratify=y, random_state=42
)

# ML Models
models = {
    'Logistic Regression': LogisticRegression(max_iter=1000),
    'Random Forest': RandomForestClassifier(),
    'SVM': SVC(probability=True),
    'XGBoost': XGBClassifier(use_label_encoder=False, eval_metric='logloss'),
    'KNN': KNeighborsClassifier(),
    'MLP': MLPClassifier(max_iter=500)
}

# CV configuration
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# Result storage
results = {}
cv_scores = {}

# Prepare ROC Curve plot
plt.figure(figsize=(10, 8))
colors = ['blue', 'green', 'red', 'orange', 'purple', 'cyan']

# Model Evaluation Loop
for i, (name, model) in enumerate(models.items()):
    print(f"\nüîç Evaluating: {name}")
    
    # Fit model
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    y_prob = model.predict_proba(X_test)[:, 1]

    # Performance
    acc = accuracy_score(y_test, y_pred)
    auc = roc_auc_score(y_test, y_prob)
    results[name] = {'Accuracy': acc, 'ROC AUC': auc}
    
    # Classification Report
    print(f"‚úÖ Accuracy: {acc:.4f}")
    print(f"‚úÖ ROC AUC: {auc:.4f}")
    print(classification_report(y_test, y_pred, target_names=class_names))

    # Confusion Matrix
    cm = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(5, 4))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=class_names, yticklabels=class_names)
    plt.title(f"{name} - Confusion Matrix")
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.tight_layout()
    plt.show()

    # ROC Curve
    fpr, tpr, _ = roc_curve(y_test, y_prob)
    plt.plot(fpr, tpr, label=f"{name} (AUC = {auc:.2f})", color=colors[i])

    # Cross-validation
    scores = cross_val_score(model, X_scaled, y, cv=cv, scoring='accuracy')
    cv_scores[name] = {'mean': scores.mean(), 'std': scores.std()}
    print(f"üîÅ Cross-Validation Accuracy: Mean = {scores.mean():.4f}, Std = {scores.std():.4f}")

# Final ROC curve
plt.plot([0, 1], [0, 1], 'k--')
plt.title("ROC Curve Comparison")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Feature importance
for name, model in models.items():
    if hasattr(model, 'feature_importances_'):
        importances = model.feature_importances_
        indices = np.argsort(importances)[-20:]
        plt.figure(figsize=(8, 6))
        plt.barh(range(len(indices)), importances[indices], align='center')
        plt.yticks(range(len(indices)), [X.columns[i] for i in indices])
        plt.title(f"{name} - Top 20 Feature Importances")
        plt.tight_layout()
        plt.show()

# Summary
print("\nüìä Final Model Accuracy & AUC:")
for name, metrics in results.items():
    print(f"{name}: Accuracy = {metrics['Accuracy']:.4f}, ROC AUC = {metrics['ROC AUC']:.4f}")

print("\nüîÅ Cross-Validation Summary (Accuracy):")
for name, scores in cv_scores.items():
    print(f"{name}: Mean = {scores['mean']:.4f}, Std = {scores['std']:.4f}")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE

# Load dataset
df = pd.read_csv('reduced_rf_dataset.csv')
X = df.drop(columns=['Label'])
y = df['Label']

# Encode labels
le = LabelEncoder()
y_encoded = le.fit_transform(y)
labels = le.classes_

# Standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# -------------------- üìâ PCA Visualization --------------------
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

plt.figure(figsize=(8, 6))
for label in np.unique(y_encoded):
    plt.scatter(
        X_pca[y_encoded == label, 0],
        X_pca[y_encoded == label, 1],
        label=labels[label], alpha=0.7
    )
plt.title("PCA - 2D Visualization of Reduced Features")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig("pca_visualization.png", dpi=300)
plt.show()

# -------------------- üîç t-SNE Visualization --------------------
tsne = TSNE(n_components=2, perplexity=30, n_iter=1000, random_state=42)
X_tsne = tsne.fit_transform(X_scaled)

plt.figure(figsize=(8, 6))
for label in np.unique(y_encoded):
    plt.scatter(
        X_tsne[y_encoded == label, 0],
        X_tsne[y_encoded == label, 1],
        label=labels[label], alpha=0.7
    )
plt.title("t-SNE - 2D Visualization of Reduced Features")
plt.xlabel("Component 1")
plt.ylabel("Component 2")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig("tsne_visualization.png", dpi=300)
plt.show()

# -------------------- üîó Correlation Heatmap --------------------
plt.figure(figsize=(12, 10))
sns.heatmap(X.corr(), annot=True, fmt=".2f", cmap="coolwarm", square=True)
plt.title("Correlation Heatmap of Features")
plt.tight_layout()
plt.savefig("correlation_heatmap.png", dpi=300)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Load dataset
df = pd.read_csv('reduced_rf_dataset.csv')
X = df.drop(columns=['Label'])  # Drop label column for correlation

# Compute correlation matrix
corr_matrix = X.corr()

# ---------- üîç Identify strong correlations ----------
threshold = 0.8
strong_corr_pairs = []

for i in range(len(corr_matrix.columns)):
    for j in range(i + 1, len(corr_matrix.columns)):
        corr_value = corr_matrix.iloc[i, j]
        if abs(corr_value) >= threshold:
            feature1 = corr_matrix.columns[i]
            feature2 = corr_matrix.columns[j]
            strong_corr_pairs.append((feature1, feature2, corr_value))

# Print strongly correlated pairs
print("üîó Strongly Correlated Feature Pairs (|correlation| ‚â• 0.8):\n")
for f1, f2, val in sorted(strong_corr_pairs, key=lambda x: -abs(x[2])):
    print(f"{f1} ‚Üî {f2} : r = {val:.2f}")

# ---------- üìä Plot Correlation Heatmap ----------
plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm", square=True,
            cbar_kws={"label": "Correlation Coefficient"})
plt.title("üîó Correlation Heatmap of Reduced RF Features")
plt.tight_layout()
plt.savefig("correlation_heatmap_highlighted.png", dpi=300)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split, cross_val_score, StratifiedKFold
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, roc_auc_score, roc_curve

from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from xgboost import XGBClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier

# Load original dataset
df = pd.read_csv('reduced_rf_dataset.csv')

# Drop correlated features
features_to_drop = ['delta_std_13', 'rms_mean', 'zcr_mean', 'mfcc_mean_1']
df_cleaned = df.drop(columns=features_to_drop)

# Save the cleaned dataset
df_cleaned.to_csv('reduced_rf_cleaned_dataset.csv', index=False)
print("‚úÖ Saved cleaned dataset as 'reduced_rf_cleaned_dataset.csv'")

# Prepare features and target
target_column = 'Label'
X = df_cleaned.drop(columns=[target_column])
y = df_cleaned[target_column]

# Encode target
le = LabelEncoder()
y = le.fit_transform(y)
class_names = le.classes_

# Standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, stratify=y, random_state=42
)

# Define models
models = {
    'Logistic Regression': LogisticRegression(max_iter=1000),
    'Random Forest': RandomForestClassifier(),
    'SVM': SVC(probability=True),
    'XGBoost': XGBClassifier(use_label_encoder=False, eval_metric='logloss'),
    'KNN': KNeighborsClassifier(),
    'MLP': MLPClassifier(max_iter=500)
}

# Prepare for ROC plotting
plt.figure(figsize=(10, 8))
colors = ['blue', 'green', 'red', 'orange', 'purple', 'cyan']
results = {}
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

for i, (name, model) in enumerate(models.items()):
    print(f"\nüîç Evaluating: {name}")
    
    # Fit model
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    y_prob = model.predict_proba(X_test)[:, 1]

    # Accuracy and AUC
    acc = accuracy_score(y_test, y_pred)
    auc = roc_auc_score(y_test, y_prob)
    results[name] = {'Accuracy': acc, 'ROC AUC': auc}

    print(f"‚úÖ Accuracy: {acc:.4f}")
    print(f"‚úÖ ROC AUC: {auc:.4f}")
    print(classification_report(y_test, y_pred, target_names=class_names))

    # Confusion Matrix
    cm = confusion_matrix(y_test, y_pred)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=class_names, yticklabels=class_names)
    plt.title(f'{name} - Confusion Matrix')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.show()

    # ROC Curve
    fpr, tpr, _ = roc_curve(y_test, y_prob)
    plt.plot(fpr, tpr, label=f"{name} (AUC = {auc:.2f})", color=colors[i])

# Final ROC Curve
plt.plot([0, 1], [0, 1], 'k--')
plt.title("ROC Curve Comparison")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Cross-validation
print("\nüîÅ 5-Fold Cross-Validation Accuracies:")
for name, model in models.items():
    scores = cross_val_score(model, X_scaled, y, cv=cv, scoring='accuracy')
    print(f"{name}: Mean = {scores.mean():.4f}, Std = {scores.std():.4f}")

# Final summary
print("\nüìä Final Accuracy Summary:")
for name, metrics in results.items():
    print(f"{name}: Accuracy = {metrics['Accuracy']:.4f}, ROC AUC = {metrics['ROC AUC']:.4f}")
```
