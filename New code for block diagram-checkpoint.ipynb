```python
import os
import librosa
import numpy as np
import pandas as pd
from tqdm import tqdm
```

```python
RIPE_DIR = "data/ripe/"
UNRIPE_DIR = "data/unripe/"
SAMPLE_RATE = 44100
DURATION = 7

def extract_features(file_path):
    y, sr = librosa.load(file_path, sr=SAMPLE_RATE, duration=DURATION)

    # MFCCs
    mfccs = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=13)
    mfcc_mean = mfccs.mean(axis=1)
    mfcc_var = mfccs.var(axis=1)

    # Spectral Centroid
    centroid = librosa.feature.spectral_centroid(y=y, sr=sr)
    centroid_mean = np.mean(centroid)
    centroid_var = np.var(centroid)

    # Spectral Rolloff
    rolloff = librosa.feature.spectral_rolloff(y=y, sr=sr)
    rolloff_mean = np.mean(rolloff)
    rolloff_var = np.var(rolloff)

    # Spectral Contrast
    contrast = librosa.feature.spectral_contrast(y=y, sr=sr)
    contrast_mean = np.mean(contrast)
    contrast_var = np.var(contrast)

    # Chroma
    chroma = librosa.feature.chroma_stft(y=y, sr=sr)
    chroma_mean = np.mean(chroma)
    chroma_var = np.var(chroma)

    # Zero Crossing Rate
    zcr = librosa.feature.zero_crossing_rate(y)
    zcr_mean = np.mean(zcr)
    zcr_var = np.var(zcr)

    # RMS Energy
    rms = librosa.feature.rms(y=y)
    rms_mean = np.mean(rms)
    rms_var = np.var(rms)

    features = list(mfcc_mean) + list(mfcc_var) + [
        centroid_mean, centroid_var,
        rolloff_mean, rolloff_var,
        contrast_mean, contrast_var,
        chroma_mean, chroma_var,
        zcr_mean, zcr_var,
        rms_mean, rms_var
    ]
    return features
```

```python
# Cell 2: Feature extraction loop
data = []
labels = []

for fname in tqdm(os.listdir(RIPE_DIR)):
    if fname.endswith(".wav"):
        fpath = os.path.join(RIPE_DIR, fname)
        features = extract_features(fpath)
        data.append(features)
        labels.append(1)  # ripe

for fname in tqdm(os.listdir(UNRIPE_DIR)):
    if fname.endswith(".wav"):
        fpath = os.path.join(UNRIPE_DIR, fname)
        features = extract_features(fpath)
        data.append(features)
        labels.append(0)  # unripe

# Cell 3: Create DataFrame
columns = [f"mfcc_mean_{i+1}" for i in range(13)] + \
          [f"mfcc_var_{i+1}" for i in range(13)] + [
              "spectral_centroid_mean", "spectral_centroid_var",
              "spectral_rolloff_mean", "spectral_rolloff_var",
              "spectral_contrast_mean", "spectral_contrast_var",
              "chroma_mean", "chroma_var",
              "zcr_mean", "zcr_var",
              "rms_mean", "rms_var"
          ]

df = pd.DataFrame(data, columns=columns)
df["label"] = labels

# Cell 4: Save as CSV
df.to_csv("cantaloupe_features.csv", index=False)
print("Saved as cantaloupe_features.csv")
```

```python
from scipy.stats import ttest_ind

# Separate ripe and unripe groups
df_ripe = df[df["label"] == 1]
df_unripe = df[df["label"] == 0]

# Perform t-tests
ttest_results = []

for feature in df.columns[:-1]:  # exclude label
    stat, p_value = ttest_ind(df_ripe[feature], df_unripe[feature], equal_var=False)
    ttest_results.append((feature, p_value))

# Convert to DataFrame
ttest_df = pd.DataFrame(ttest_results, columns=["Feature", "p_value"])
ttest_df["Significant"] = ttest_df["p_value"] < 0.05  # mark significance

# Show significant features
print("Significantly different features (p < 0.05):")
print(ttest_df[ttest_df["Significant"] == True])

# Save results
ttest_df.to_csv("ttest_results.csv", index=False)
```

```python
# Cell 5: PCA analysis
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Separate features and label
X = df.drop("label", axis=1)
y = df["label"]

# Step 1: Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Step 2–4: Apply PCA
pca = PCA()
X_pca = pca.fit_transform(X_scaled)

# Step 5: Explained variance ratio
explained_variance = pca.explained_variance_ratio_
cumulative_variance = np.cumsum(explained_variance)

# Plot cumulative explained variance
plt.figure(figsize=(10, 5))
plt.plot(range(1, len(cumulative_variance)+1), cumulative_variance, marker='o')
plt.axhline(y=0.95, color='r', linestyle='--')
plt.xlabel("Number of Principal Components")
plt.ylabel("Cumulative Explained Variance")
plt.title("PCA - Cumulative Explained Variance")
plt.grid(True)
plt.show()
```

```python
from statsmodels.multivariate.manova import MANOVA

# Prepare data
X = df.drop(columns=["label"])
y = df["label"]

# Combine features and label into one DataFrame
df_manova = X.copy()
df_manova["label"] = y

# Fit MANOVA
maov = MANOVA.from_formula(' + '.join(X.columns) + ' ~ label', data=df_manova)
print(maov.mv_test())
```
